Feature Selection Part

```python
fooditem.columns
```

```python

fooditem=fooditem[['FoodCategory', 'FoodItem' , 'Cals_per100grams']]
```

```python
fooditem
```

```python
fooditem['tags']=fooditem['FoodItem']+fooditem['Cals_per100grams']
```

```python
fooditem
```

```python
new_data=fooditem.drop(columns=['FoodItem','Cals_per100grams'])
```

```python
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=2225,stop_words='english')
```

```python
cv
```

```python
vector=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['FoodCategory']=="Spreads"].index[0]
```

```python
distance=sorted(list(enumerate(similarity[9])),reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_data.iloc[i[0]].FoodCategory)
```

```python
def recommand(fooditem):
    index=new_data[new_data['FoodCategory']==fooditem].index[0]
    distance=sorted(list(enumerate(similarity[index])),reverse=True, key=lambda vector:vector[1])
    for i in distance[0:9]:
        print(new_data.iloc[i[0]].FoodCategory)
    
```

```python
recommand("Fruits")
```
